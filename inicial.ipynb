```python
#2. Análise Exploratória de Dados (EDA):
import pandas as pd
import numpy as np 

```

```python
tabela = pd.read_excel("dados_wgs.xlsx")
tabela = tabela.drop(columns="id")
tabela = tabela.dropna()
display(tabela)
display(tabela.info())
```

```python
from sklearn.preprocessing import LabelEncoder

codificador = LabelEncoder()
for coluna in tabela.columns:
    if tabela[coluna].dtype == "object":
        tabela[coluna] = codificador.fit_transform(tabela[coluna])

display(tabela.info())

```

```python
x = tabela.drop(["valor_total","link","qtd_dados_bairro"], axis=1)
y = tabela["valor_total"]

from sklearn.model_selection import train_test_split
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression


modelo_linear_regression = LinearRegression()
```

```python
#Aplicando o feature scaling com MMS e Treinando Modelo
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
x_train_scaled = scaler.fit_transform(x_treino)
x_test_scaled = scaler.transform(x_teste)

modelo_linear_regression.fit(x_train_scaled, y_treino)
```

```python
#Avaliação e previsões
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

previsao_linear = modelo_linear_regression.predict(x_test_scaled)

mse = mean_squared_error(y_teste, previsao_linear)
print(f'Mean Squared Error (MSE:) {mse}')

mae = mean_absolute_error(y_teste, previsao_linear)
print(f'Mean Absolute Error (MAE): {mae}')

r2 = r2_score(y_teste, previsao_linear)
print(f'R² Score: {r2}')
```
